plot sin(x) with noise

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,1.05,0.05)
y=np.sin(2*np.pi*x)
y1=y+np.random.normal(scale=0.1,size=y.shape)
plt.plot(x,y,'rs--')
plt.plot(x,y1,'b^:')
plt.show()
```

identifying signal monotonicity

```python
def mon_det(x):
    inc=1
    dec=1
    prev=x[0]
    for i in range(1,len(x)):
        if x[i]>=prev:
            inc +=1
        if x[i]<=prev:
            dec +=1
        if inc!=(i+1) and dec!= (i+1):
            break
        prev=x[i]
    if inc==len(x):
        print(f'your array {x} is monotonic (increasing)')
        out=(True,'increasing')
    elif dec==len(x):
        print(f'your array {x} is monotonic (decrasing)')
        out=(True,'decreasing')
    else:
        print(f'your array {x} is not monotonic')
        out=(False,None)
    return out
x=[]
inp=input('enter your array (seperate values by space)').split()
for i in inp:
    x.append(float(i))

y=mon_det(x)
```
